3.	Write a program that can determine the distance between two points given the coordinates using the formula:

Hint/Rule: No library or package is needed to implement this equation. Example: x2, y2 = -3, 3 and x1, y1 = 2, 2 d = 5.099019514

```python
x1, y1 = map(float, input("Enter coordinates x1 y1: ").split())
x2, y2 = map(float, input("Enter coordinates x2 y2: ").split())

distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5
print("Distance between points =", distance)
```
